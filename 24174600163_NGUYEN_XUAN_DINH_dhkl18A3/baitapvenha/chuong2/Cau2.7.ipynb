```python
import json

def json_to_python(json_data):
    return json.loads(json_data)

if __name__ == "__main__":
    json_data = '{"name": "Tona", "age": 18, "city": "HN"}'
    python_object = json_to_python(json_data)

    print("Đối tượng Python sau khi chuyển đổi:")
    print(python_object)
```
