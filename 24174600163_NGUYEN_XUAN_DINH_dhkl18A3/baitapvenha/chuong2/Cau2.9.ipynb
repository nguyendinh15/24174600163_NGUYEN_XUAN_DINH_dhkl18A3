```python
import json

if __name__ == "__main__":
    python_dict = {
        "name": "Tona",
        "age": 18,
        "city": "HN",
        "is_student": False
    }
    json_string = json.dumps(python_dict, indent=4, sort_keys=True)
    # In ra các thành viên với mức thụt lề 4
    print(f"Chuỗi JSON sau khi chuyển đổi:\n{json_string}")
```
