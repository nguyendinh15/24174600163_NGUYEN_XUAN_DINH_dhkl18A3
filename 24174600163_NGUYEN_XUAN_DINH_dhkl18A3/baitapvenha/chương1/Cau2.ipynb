```python
class TS:
    def __init__(self, ho_ten = "", toan = 0, ly = 0, hoa = 0):
        self.ho_ten = ho_ten
        self.toan = toan
        self.ly = ly
        self.hoa = hoa

    def nhap(self):
        self.ho_ten = input("Nhap ho va ten thi sinh: ")
        self.toan = float(input("Nhap diem Toan: "))
        self.ly = float(input("Nhap diem Ly: "))
        self.hoa = float(input("Nhap diem hoa: "))

    def tong_diem(self):
        return self.toan + self.ly + self.hoa

    def display(self):
        print("Ho va ten:", self.ho_ten)
        print("Diem Toan:", self.toan)
        print("Diem Ly:", self.ly)
        print("Diem Hoa:", self.hoa)
        print("Tong diem:", self.tong_diem())
        print("-------------------------")

ds = []
n = int(input("Nhap so luong thi sinh: "))
for i in range(n):
    ts = TS()
    ts.nhap()
    ds.append(ts)    

#Sắp xếp tổng điểm giảm dần
ds.sort(key=lambda x: x.tong_diem(), reverse = True)

print(" -----Danh sach thi sinh trung tuyen-----")
for ts in ds:
    if ts.tong_diem() >= 20:
        ts.display()
```
