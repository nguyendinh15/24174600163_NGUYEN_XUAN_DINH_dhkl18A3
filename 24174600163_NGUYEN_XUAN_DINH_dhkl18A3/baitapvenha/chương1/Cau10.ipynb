```python
import math
class Diem:
    def __init__(self, x, y):
        self.x = x
        self.y = y

class Elip(Diem):
    def __init__(self, x_tam, y_tam, a_do_dai_truc, b_do_dai_truc):
        super().__init__(x_tam, y_tam)
        self.a = a_do_dai_truc
        self.b = b_do_dai_truc

    def dien_tich(self):
        return math.pi * self.a * self.b

class duong_tron(Elip):
    def __init__(self, x_tam, y_tam, ban_kinh):
        super().__init__(x_tam, y_tam, ban_kinh, ban_kinh)

elip = Elip(0, 0, 7, 4)
elip_dien_tich = elip.dien_tich()
print("Diện tích của elip là:", elip_dien_tich)

duong_tron = duong_tron(0, 0, 5)
duong_tron_dien_tich = duong_tron.dien_tich()
print("Diện tích của hình tròn là:", duong_tron_dien_tich)

```
