```python
class Stack:
    def __init__(self):
        self.stack = []

    def push(self, item):
        self.stack.append(item)

    def pop(self):
        if not self.is_empty():
            return self.stack.pop()
        else:
            return None

    def is_empty(self):
        return len(self.stack) == 0

    def count(self):
        return len(self.stack)

    def print(self):
        if self.is_empty():
            print("Ngăn xếp rỗng.")
        else:
            print("Nội dung ngăn xếp:")
            for item in self.stack:
                print(item)

stack = Stack()
stack.push(10)
stack.push(20)
stack.push(30)

stack.print()
print("Số phần tử:", stack.count())
```
