```python
import math
class Da_giac:
    def __init__(self, so_canh):
        self.so_canh = so_canh
        self.do_dai = []

    def nhap_do_dai(self):
        for i in range(self.so_canh):
            canh = float(input(f"Nhap do dai canh {i+1}: "))
            self.do_dai.append(canh)

    def chu_vi(self):
        return sum(self.do_dai)

class HBH(Da_giac):
    def __init__(self):
        super().__init__(4) 

    def dien_tich(self):
        day = self.do_dai[0]
        cao = float(input("Nhap chieu cao: "))
        return day * cao

class HCN(HBH):
    def __init__(self):
        super().__init__()  

class HV(HCN): 
    def __init__(self): 
      super().__init__() 
      
HV = HV()
HV.nhap_do_dai()
chu_vi_HV = HV.chu_vi()
dien_tich_HV = HV.dien_tich()
print(f"Chu vi hinh vuong la: {chu_vi_HV}")
print(f"Dien tich hinh vuong la: {dien_tich_HV}")
```
