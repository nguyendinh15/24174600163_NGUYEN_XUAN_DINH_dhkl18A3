```python
class PS:
    def __init__(self, tuso = 0, mauso = 1):
        self.tuso = tuso
        self.mauso = mauso

    def nhap(self):
        self.tuso = int(input("Nhap vao tu so: "))
        self.mauso = int(input("Nhap vao mau so (khac 0): "))

    def ktra_hop_le(self):
        return self.mauso != 0
    
    def display(self):
        print(f"Phan so la: {self.tuso}/{self.mauso}")

ps = PS()
ps.nhap()
if ps.ktra_hop_le():
    ps.display()
    print("Phan so hop le")
else:
    print("Phan so khong hop le")

```
