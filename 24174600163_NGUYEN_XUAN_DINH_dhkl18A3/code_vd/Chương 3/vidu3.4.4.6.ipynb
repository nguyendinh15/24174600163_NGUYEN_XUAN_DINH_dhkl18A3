```python

import numpy as np
#Tạo màng dữ liệu với kích thước lớn (1000 phân từ)
data = np. random. randint(100, 200, size=1000)
# Tạo một giá trị trung bình (sô thực)
mean_value = np.array([data.mean()])
# Tính tỷ lệ phần trăm tăng trường băng cách áp dụng broadcasting 
percentage_growth= (data - mean_value) / mean_value * 100
# In ra 10 phân tử đâu tiên của mảng dữ liệu và tỷ lệ phân trăm tăng trường 
print( "Dữ liệu gộc:", data[:10])
print( "Tỷ lệ phân trăm tăng trường:", percentage_growth[:10])
```
