```python
import numpy as np
arr = np.array([10, 20, 30, 40, 50, 60, 70, 80, 90]) 
 
#Truy xuất các phần tử bằng một danh sách các chỉ mục 
indices = [1, 3, 4] 
sub_arr_1 = arr[indices] 
 
print(sub_arr_1) 
```
