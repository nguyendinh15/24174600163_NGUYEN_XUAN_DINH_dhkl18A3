```python
import numpy as np
# Tạo mảng ngẫu nhiên 
arr1 = np.array([1, 2, 3, 4, 5]) 
arr2 = np.array([5, 4, 3, 2, 1]) 
 
# Toán tử so sánh dạng ufuncs 
result1 = arr1 > arr2   # Lớn hơn 
result2 = arr1 == arr2  # Bằng nhau 
result3 = arr1 <= arr2  # Nhỏ hơn hoặc bằng 
 
print("Kết quả lớn hơn:", result1)     
print("Kết quả bằng nhau:", result2)   
print("Kết quả nhỏ hơn hoặc bằng:", result3)  
```
