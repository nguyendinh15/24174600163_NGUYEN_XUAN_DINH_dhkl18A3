```python
import numpy as np 
arr = np.array([[1,0,3], 
                [0,5,0], 
                [7,0,9]]) 
#Sử dụng phương thức sum kèm theo tham số axis (trục) tương ứng 
print("Số phần tử khác 0 theo mỗi hàng:", np.sum(arr !=0, axis=1)) 
print("Số phần tử khác 0 theo mỗi cột:",np.sum(arr!=0, axis=0)) 
```

```python
print("Trong mảng arr có phần tử > 5: ", np.any(arr > 5)) 
print("Trong mảng arr có phần tử < 0: ", np.any(arr < 0)) 
print("Trong mảng arr tất cả các phần tử đều < 10: ", np.all(arr < 10)) 
print("Trong mảng arr tất cả các phần tử đều = 3:", np.all(arr == 3)) 
```

```python
print("Kiểm tra từng cột trong mảng arr, \ncột nào có tất cả các phần tử lớn hơn 1: \n",  
      np.all(arr > 1, axis=0)) 
```
