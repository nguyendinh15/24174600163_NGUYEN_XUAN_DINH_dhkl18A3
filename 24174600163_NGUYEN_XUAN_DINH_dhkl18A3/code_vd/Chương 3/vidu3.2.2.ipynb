```python
import numpy as np 
 
#Kiểu dữ liệu nguyên: 
arr_int=np.array([1,2,3,4],dtype=np.int32) 
arr_int.dtype 
```

```python
arr_float = np.array([1.1, 2.2, 3.3], dtype=np.float64) 
arr_float.dtype
```

```python
arr_string = np.array(['apple', 'banana', 'orange'], 
dtype=np.str_) 
arr_string.dtype
```

```python
arr_bool = np.array([True, False, True], dtype=np.bool_) 
arr_bool.dtype 
```
