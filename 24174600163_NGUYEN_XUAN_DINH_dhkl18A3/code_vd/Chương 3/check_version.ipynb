```python
import numpy as np  
import sys  

S = list(range(1000)) 
items_S= sys.getsizeof(S) 
print("Kích thước mỗi phẫn tử trong List tính theo bytes: ",items_S) 
S_size = items_S*len(S) 
print("Kích thước của toàn bộ List tính theo bytes: ",S_size) 
D= np.arange(1000)  
print("Kích thước mỗi phần tử của mảng Numpy theo bytes: ",D.itemsize) 
array_size=D.itemsize*D.size 

print("Kích thước của toàn bộ mảng Numpy array in bytes: ",array_size) 
print(f"Tỷ lệ (list/array): {S_size / array_size:.2f}")
```

```python
import numpy as np

arr = np.array ([[1,2,31], [4,5,6]]) # shape=(2,3)
print("Số chiều của máng là:", arr.ndim) 
print("Kích thước của mảng là:", arr. shape) 
print( "Tổng số phần tử trong mắng là:", arr.size) 
print ("Kiểu dữ liệu của các phần tử trong mảng là:", arr.dtype) 
print("Kích thước theo (byte) của mỗi phần tứ trong mảng là:", arr.itemsize) 
print( "Tổng kích thước theo (byte) của mảng là:", arr.nbytes)
# BỔ SUNG:
print ("Strides (bước nhảy bộ nhớ) là:", arr.strides)
 # Ví dụ theo trục (axis):
print("Tổng theo trục 0 (theo cột):", arr.sum(axis=0)) # [1+4, 2+5, 3+6]
print("Tổng theo trục 1 (theo hàng):", arr. sum(axis=1)) # [1+2+3, 4+5+6]
```
