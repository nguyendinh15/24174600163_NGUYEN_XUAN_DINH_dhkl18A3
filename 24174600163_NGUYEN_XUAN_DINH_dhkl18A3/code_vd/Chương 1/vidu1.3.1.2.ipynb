```python
# Ví dụ 1.3.1.2. Minh họa sử dụng setter, getter và deleter

class Dog:
    def __init__(self, name, size, age, color):
        self._name = name      # Thuộc tính protected
        self._size = size      # Thuộc tính protected
        self.__age = age       # Thuộc tính private
        self.__color = color   # Thuộc tính private

    @property # Định nghĩa phương thức getter cho thuộc tính private __age
    def age(self):
        print("Getter for age called")
        if not hasattr(self, "_Dog__age"):   # Kiểm tra xem __age có tồn tại không
            raise AttributeError("Age has been deleted or not set")
        return self.__age     # Trả về giá trị tuổi (private)

    @age.setter
    def age(self, value):
        """Setter cho thuộc tính private __age"""
        print("Setter for age called")
        if value < 0:
            raise ValueError("Age cannot be negative")
        self.__age = value    # Cập nhật giá trị tuổi (private)

    @age.deleter
    def age(self):
        """Định nghĩa delete cho thuộc tính private __age"""
        print("Deleter for age called")
        if not hasattr(self, "_Dog__age"):   # Kiểm tra trước khi xóa
            raise AttributeError("Age has already been deleted or not set")
        del self.__age        # Xóa thuộc tính private __age

dog =Dog("Bull", "Large", 2, "yellow")
print(f"Bull's age is: {dog.age}") #Truy cap thong qua getter

dog.age = 3
print(f"Bull updated Age is: {dog.age}")

del dog.age 
```
