```python
import pandas as pd 

indA = pd.Index([0, 2, 4, 6, 8]) 
indB = pd.Index([1, 3, 7, 8, 9]) 

print('Index A:', indA) 
print('Index B:', indB) 

print('Hợp:', indA.union(indB))  # union 

print('Hiệu A - B:', indA.difference(indB))  # difference

print('Giao:', indA.intersection(indB))  # intersection

print('Hiệu đối xứng:', indA.symmetric_difference(indB))  # symmetric difference

# Kiểm tra A là tập con của B?
print('A thuộc B:', indA.isin(indB))
```

```python
import pandas as pd

dates = pd.date_range('2025-01-01', periods=5, freq='D')  # Tạo DatetimeIndex
s = pd.Series([1, 2, 3, 4, 5], index=dates)

print(type(s.index).__name__)
```

```python
print(s.loc['2025-01-03'])   # Truy cập theo ngày

print('\nCắt lát theo tháng:')
print(s.loc['2025-01'])      # Lấy toàn bộ tháng 01/2025
```

```python
import pandas as pd

# Dữ liệu
tinh = ['Hà Nội', 'Hà Nội', 'Đà Nẵng', 'Đà Nẵng']
nam = [2023, 2024, 2023, 2024]

# Tạo MultiIndex
idx = pd.MultiIndex.from_arrays([tinh, nam], names=['Tinh/TP', 'Năm'])

# DataFrame
df = pd.DataFrame({
    'Dân số': [8330830, 8410000, 1191381, 1210000],
    'Diện tích': [3358.9, 3358.9, 1284.9, 1284.9]
}, index=idx)

print(df)

print("\nTruy cập (Hà Nội, 2024):")
print(df.loc[('Hà Nội', 2024), 'Dân số'])

print("\nMọi năm của Đà Nẵng:")
print(df.loc[('Đà Nẵng', slice(None)), :])
```
