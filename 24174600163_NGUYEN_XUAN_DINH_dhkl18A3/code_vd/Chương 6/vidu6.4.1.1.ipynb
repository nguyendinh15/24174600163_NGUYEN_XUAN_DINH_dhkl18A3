```python
import threading 
import time
# global variable x 
x = 0 
   
#Viết hàm increment(), tăng giá trị biến toàn cục x 
def increment():     
    global x 
    temp = x
    time.sleep(0.000001)
    x = temp + 1 
#Hàm theread_task() thực hiện tác vụ luồng,  thực hiện gọi hàm  
# increment() 100000 lần.''' 
def thread_task():                
    for _ in range(100000): 
        increment() 
  
#Hàm main_task(): thực hiện nhiệm vụ chính của chương trình.  
#Nó đặt giá trị của biến toàn cục x=0, tạo hai luồng (t1 và t2)  
# để thực thi công việc trong thread_task(). 
def main_task(): 
    global x 
    x = 0 
   
    # creating threads 
    t1 = threading.Thread(target=thread_task) 
    t2 = threading.Thread(target=thread_task) 
   
    # start threads 
    t1.start() 
    t2.start() 
   
    # wait until threads finish their job 
    t1.join() 
    t2.join() 
#Chương trình chạy vòng lặp 10 lần,  
#trong mỗi lần lặp gọi main_task() để thực hiện công việc. 
#Sau đó in giá trị x ở mỗi lần lặp ra màn hình. 
for i in range(10): 
    main_task() 
    print("Lần lặp {0}: x = {1}".format(i, x)) 
```

```python
import threading
import time

balance = 100      # Biến toàn cục, được CHIA SẺ giữa các thread

def add_money():
    global balance
    tmp = balance        # Bước 1: đọc số dư hiện tại vào biến tạm
    time.sleep(0.001)    # Bước 2: ngủ 1ms để có tính tạo cơ hội cho thread khác can thiệp
    balance = tmp + 10   # Bước 3: viết giá trị đã cập nhật trở lại vào balance

# Tiếp theo: tạo 2 thread, cả 2 đều chạy cùng một hàm add_money()
threads = [threading.Thread(target=add_money) for _ in range(2)]

# Cả 2 thread bắt đầu chạy gần như cùng lúc
for t in threads:
    t.start()

# Đợi cả 2 thread chạy xong
for t in threads:
    t.join()

print("Số dư cuối:", balance)
```
