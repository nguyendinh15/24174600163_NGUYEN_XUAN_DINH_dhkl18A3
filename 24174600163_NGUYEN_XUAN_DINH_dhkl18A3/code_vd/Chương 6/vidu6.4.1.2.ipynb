```python
import threading

# Biến toàn cục
x = 0

# Hàm tăng x
def increment():
    global x
    x += 1

# Tác vụ của mỗi thread
def thread_task(lock):
    for _ in range(100000):
        lock.acquire()
        increment()
        lock.release()

# Tác vụ chính
def main_task():
    global x
    x = 0

    # Tạo lock
    lock = threading.Lock()

    # Tạo 2 luồng
    t1 = threading.Thread(target=thread_task, args=(lock,))
    t2 = threading.Thread(target=thread_task, args=(lock,))

    # Chạy luồng
    t1.start()
    t2.start()

    # Đợi luồng xong
    t1.join()
    t2.join()


# Chạy 10 lần
for i in range(10):
    main_task()
    print("Lần lặp {0}: x = {1}".format(i, x))
```
